### Q2. Swap Two Variables

```python
a = 4
b= 5
print('Enter first number: ', a)
print('Enter second number: ',b)
c = a
a = b
b = c
print("\nafter swapping:")
print('Enter first number: ',a)
print('Enter second number: ', b)
```

### Q3. Swap Two Variables


```python
x = 4
y= 5
print('\nEnter first number: ', x)
print('Enter second number: ',y)
x = x + y
y = x - y
x = x - y
print("\nafter swapping:")
print('Enter first number: ',x)
print('Enter second number: ', y)
```

### Q4. Convert Temperature from Celsius to Fahrenheit


```python
# Q4. Convert Temperature from Celsius to Fahrenheit
clesius = 25
print('\nEnter temperature in Celsius: ',clesius)
F = 9/5 * clesius + 32
print('Temperature in Fahrenheit: ' ,F)
```
