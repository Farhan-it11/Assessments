# 1. Student Grade Calculator
###### Write a function calculate_grade(marks) that takes marks as input and returns grade according to:
###### 90 and above → A
###### 80 – 89 → B
###### 70 – 79 → C
###### Below 70 → Fail

```python
def calculate_grade(marks):
    
    if marks <=100 and marks >=90:
        print('You have A Grade')
    elif marks <=89 and marks >=80:
        print('You have B Grade')
    elif marks <=79 and marks >=70:
        print('You have C Grade')
    else:
        print('You failed')
```

```python
marks = int(input("Eneter number:"))
calculate_grade(marks)
```

# 2. Write a function is_even(number) that returns:
###### True if the number is even
###### False if the number is odd

```python
def is_even(number):
    for x in range(2, number):
        if number % x == 0:
            return print (True, ":Number is even.")
        else:
            return print (False, ":Number is odd.")
```

```python
is_even(8)
```

# 3. Factorial Using Function
###### Write a function factorial(n) that returns the factorial of a number using a loop.
###### Example:
###### factorial(5)
###### Output → 120

```python
def factorial(n):
    fac = 1
    for x in range(n, 0, -1):
        fac = fac * x
    return fac
```

```python
factorial(10)
```

# 4. Find Maximum Number
###### Write a function find_max(a, b, c) that returns the largest number among three numbers.

```python
def find_max(a, b, c):
    if a > b and a > c:
        print(f'{a} is the largest number.')
    elif b > a and b > c:
        print(f'{b} is the largest number.')
    else:
        print(f'{c} is the largest number.')
```

```python
find_max(7, 40, 5)
```

# 6. Prime Number Checker
###### Write a function is_prime(n) that checks whether a number is a prime number.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
n = 3
if is_prime(n):
    print(n, "is a prime number")
else:
    print(n, "is not a prime number")
```

# 7. List Sum Function
###### Write a function list_sum(numbers) that takes a list as input and returns the sum of all numbers.
###### Example:
###### [10, 20, 30]
###### Output → 60

```python
def list_sum(numbers):
    add = 0
    for x in numbers:
        add += x
    return add
```

```python
numbers = [10, 5, 20]
list_sum(numbers)
```

# 8. Password Strength Checker
###### Write a function check_password(password) that checks:
###### Minimum 8 characters required
###### Must contain at least one digit
###### Return:
###### "Strong" or "Weak"

```python
def check_password(password):
    if len(password) >= 8:
        if password.isalnum() == True and password.isdigit() == False and password.isalpha() == False:
            return print("Strong")
        else:
            return print("Weak")
    else:
        return print("Weak")
```

```python
check_password("password1")
```

# 9. Temperature Converter

# 10. Remove Duplicates from List
###### Write a function remove_duplicates(data) that removes duplicate elements from a list and returns a new list.
###### Example:
###### [1,2,2,3,4,4]
###### Output → [1,2,3,4]

```python
def remove_duplicates(data):
    i = 0
    while i < len(data):
        j = i + 1
        
        while j < len(data):
            if data[i] == data[j]:
                data.pop(j)
            else:
                j += 1
        
        i += 1
    
    return sorted(data)
```

```python
data = [1,2,2,3,5,4,4,2,5,2]
remove_duplicates(data)
```
