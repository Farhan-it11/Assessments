# Part 1: Working with Variables, Input, and Strings

### 1. Prompt for user input

```python
full_name = input("Enter your full name:  ")
print(f"Hello, {full_name}! Welcome to the Python assignment.")
```

### 2. String methods

```python
print(f'Your name in uppercase: {full_name.upper()}')
print(f'Your name in lowercase: {full_name.lower()}')
print('Length of your name: ',len(full_name))
name = full_name.split()
print('Name with hyphens: ','-'.join(name))
full_name = full_name.strip()
```

### 3. Type casting and numeric operations

```python
birth_year = int(input("ENTER YOUR BIRTH YEAR: "))
age = 2026 - birth_year
print('You are approximately',age,'years old.')
```

# Part 2: Lists and Indexing

### 1. Create a list from user input

```python
fruit_name = input('Enter three of your favorite fruits (comma-separated): ')
fruit_list = fruit_name.split(',')
print('Your list of fruits: ',fruit_list)
```

### 2. List indexing and slicing

```python
print('First fruit:',fruit_list[0])
print('Last fruit:',fruit_list[2])
print('Second fruit:',fruit_list[1])
print('First two fruits:', fruit_list[0: 2])
```

### 3. List methods

```python
fruit_list.append('orannge')
print('''Adding a new fruit...
Updated list: ''', fruit_list)
```

```python
fruit_list.remove('mango')
print('''Removing one fruit...
Updated list: ''', fruit_list)
```

```python
print('Length of the list is now ',len(fruit_list))
```

# Part 3 (Optional Challenge): Combining Concepts

### 1. String & List Interaction

```python
letters_list = list(full_name.replace(" ", ""))
print(f'''Creating a letters list from your full name...
First 3 letters: {letters_list[0: 3]}
Last 3 letters: {letters_list[-3:]}''')
```

### 2. Manipulate and Reconstruct

```python
first_three_letters = ''.join(letters_list[0: 3]).upper()
rest_of_letters = ''.join(letters_list[4:]).lower()
modified_name = first_three_letters + rest_of_letters
print (f'''Reconstructing name with some modifications...
Modified name: {modified_name}''')
```

### 3. Simple Arithmetic on a List of Numbers

```python
numbers = input("Enter 5 numbers (space-separated): ")
x = numbers.split()
numbers_list = list([int(x[0]), int(x[1]), int(x[2]), int(x[3]), int(x[4])])
total_sum = numbers_list[0] + numbers_list[1] + numbers_list[2] + numbers_list[3] + numbers_list[4]
print('Sum of your numbers: ', total_sum)
average = total_sum / len(numbers_list)
print('Average of your numbers: ', average)
```
