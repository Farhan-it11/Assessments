# 3. Write a program to calculate the sum of numbers from 1 to N entered by the user.

```python
num = int(input("Enter number: "))
i = 1
Sum = 0
while i <= num:
    Sum += i
    i = i + 1
print ('Sum of numbers is', Sum)
```

# 4. Print the multiplication table of a number entered by the user (up to 10).


```python
num = int(input('Eneter number : '))
i = 1
while i <= 10:
    print(f'{num} * {i} = {num * i}')
    i = i + 1
```

# 5. Write a program to reverse a string using a for loop.

```python
string = input("Enter a string: ")

rev_text = ""
for i in range(-1 , -(len(string)+1) , -1):
    rev_text += string[i]

print("Reversed string:", rev_text)
```

# 6. Count the number of vowels present in a string using a for loop.

```python
string = "My name is Muhammad Farhan Malik"
vowel_count = 0
vowels = "aeiouAEIOU"

for char in string:
    if char in vowels:
        vowel_count += 1
print("Number of vowels:", vowel_count)
```

# 8. Write a program to check whether a number is prime using a for loop.
## Given a list:
## numbers = [12, 5, 7, 18, 21, 30]

```python
numbers = [12, 5, 7, 18, 21, 30]

for x in numbers:
        for i in range(2, x):
            if x % i == 0:
                print(x, "is not a Prime number")
                break
        else:
            print(x, "is a Prime number")
           
```

# 10. Break and Continue
## Print numbers from 1 to 20 but: - Skip multiples of 3. - Stop execution if number becomes greater than 15.


```python
for num in range(0, 20):
    if num > 15:
        break
    if num % 3 == 0:
        continue
    print(num)
```

# 11. Calculate factorial of a number using a while loop.

```python
num = 15
factorial = 1

for i in range(num, 0, -1):
    factorial *= i
print(f"Factorial of {num} is {factorial}")
```
